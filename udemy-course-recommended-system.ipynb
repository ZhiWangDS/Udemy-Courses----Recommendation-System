## Popularity Based Recommendation System

Popularity scores have define by number of subscribers and number of reviews 

```python
data.columns
```

```python
def pop_rem(data):
    # Create a column for popularity score
    data['pop_score'] = 0.6*data['num_subscribers']+0.4*data['num_reviews']    
    #create a DataFrame with the sorted new column by descending order
    df_sorted = data.sort_values(by ='pop_score',ascending = False)
    return df_sorted
```

```python
pop_rem(data).head(3)
```

```python
# we are interested in 'course_title' and 'pop_score'
pop_rem(data)[['course_title','pop_score']].head(5)
```

```python
import neattext.functions as nfx
data['course_title']
```

```python
# remove stop words and special characters from course titles, 
# which is allowing more focus on important content, and reduce the new dataset size

data['course_title'] = data['course_title'].apply(nfx.remove_stopwords)
data['course_title']= data['course_title'].apply(nfx.remove_special_characters)
data['course_title']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# try top 3000 most frequent words in corpus(collection of all course titles)
cv = CountVectorizer(max_features = 3000)
vectors = cv.fit_transform(data['course_title']).toarray()
vectors
```

```python
vectors.shape,len(cv.get_feature_names_out())
```

```python
#Cosine similarity is metric used to measure the similariu between two vectors in a multi-dimensional space 
# by calculating the cosine of the angle between the two vectors. 
#EX: A[1,1,1],B[0,1,0] => 
   #cosine similarity(A,B) = (A*B)/(||A||*||B||) = (1*0 + 1*1+ 1*0)/(sqrt(1^2+1^2+1^2)*sqrt(0^2+1^2+0^2)

from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
similarity # diagnal always 1
```

```python
similarity.shape
```

```python
# sorting the similatiry with index using the enumerate function, let see top 5
sorted(enumerate(similarity[0]),reverse = True,key = lambda x:x[1])[1:6]
```

```python
data.iloc[39]['course_title']
```

```python
data[data['course_title']== "Complete Investment Banking Course 2017"].index[0]
```

```python
def content_recomm(course_content):
    course_index =data[data['course_title']==course_content].index[0]
    sim = similarity[course_index]
    course_list = sorted(enumerate(sim),reverse = True,key = lambda x:x[1])[1:6]
    
    for i in course_list:
        print(data.iloc[i[0]]['course_title'])
    
    
```

```python
content_recomm("Ultimate Investment Banking Course")
```
